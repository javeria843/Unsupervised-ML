```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=23)

# Check the shape of the data
print(X.shape)
```

```python
# Visualize the data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=30)
plt.title("Synthetic Data with 3 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42
)
```

```python
from sklearn.cluster import KMeans
```

ELBOW METHOD

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Elbow Method to find optimal K
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# Plot the Elbow Curve
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means')
```

```python
from sklearn.cluster import KMeans

# Apply KMeans with 3 clusters
kmeans = KMeans(n_clusters=3, init="k-means++", random_state=42)

# Predict cluster labels for training data
y_labels = kmeans.fit_predict(X_train)

# Output cluster labels
print(y_labels)
```

```python
y_test_label= kmeans.predict(X_test)
```

```python
import matplotlib.pyplot as plt

# Visualize clustered training data
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_labels, cmap='viridis', s=30)
plt.title("KMeans Clustering on Training Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Visualize clustered training data
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test_label, cmap='viridis', s=30)
plt.title("KMeans Clustering on Training Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

**Silhouette Score**

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Silhouette Score for K = 2 to 10
silhouette_coefficients = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", random_state=42)
    kmeans.fit(X_train)
    score = silhouette_score(X_train, kmeans.labels_)
    silhouette_coefficients.append(score)

# Print the scores
print("Silhouette Scores:", silhouette_coefficients)
```

```python
plt.plot(range(2, 11), silhouette_coefficients, marker='o')
plt.xticks(range(2, 11),silhouette_coefficients)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title
```

**HIRARICAL CLUSTERING**

```python
# Hierarchical Clustering Implementation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = datasets.load_iris()

# Convert to DataFrame
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris_data)

# Show the standardized data (optional)
print(pd.DataFrame(scaled_data, columns=iris.feature_names).head())
```

```python
from sklearn.preprocessing import StandardScaler

# Create the scaler object
scaler = StandardScaler()

# Apply standardization to the iris dataset
X_scaled = scaler.fit_transform(iris_data)

# Show the standardized data (optional)
print(X_scaled)

# Check the shape of the scaled data
print(X_scaled.shape)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca
```

```python
pca_scaled=pca.fit_transform(X_scaled)
```

```python
# Dendrogram banane ke liye import
import scipy.cluster.hierarchy as sc
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

# Dendrogram plot karna
plt.figure(figsize=(20, 7))
plt.title("Dendrogram")
sc.dendrogram(sc.linkage(pca_scaled, method='ward'))
plt.xlabel("Sample Index")
plt.ylabel("Euclidean Distance")
plt.grid(True)
plt.show()

```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Apply Agglomerative Clustering (no need for 'affinity' when using 'ward')
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
cluster.fit(pca_scaled)

# Cluster labels
print(cluster.labels_)

# Visualize the clusters
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c=cluster.labels_, cmap='viridis', s=30)
plt.title("Agglomerative Clustering Results")
plt.xlabel("PCA Feature 1")
plt.ylabel("PCA Feature 2")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Silhouette Scores for K = 2 to 10
silhouette_coefficients = []

for k in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=k, linkage='ward')
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_coefficients.append(score)

# Plotting Silhouette Scores
plt.plot(range(2, 11), silhouette_coefficients, marker='o')
plt.xticks(range(2, 11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Score vs Number of Clusters')
plt.grid(True)
plt.show()
```

DBSCAN Clusterring

```python
# Import libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline

# Generate synthetic moon-shaped data
X, y = make_moons(n_samples=250, noise=0.05)

# Visualize raw data
plt.scatter(X[:, 0], X[:, 1])
plt.title("Raw Moon-Shaped Data")
plt.grid(True)
plt.show()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(X_scaled)

# Visualize DBSCAN clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan.labels_, cmap='plasma', s=30)
plt.title("DBSCAN Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
